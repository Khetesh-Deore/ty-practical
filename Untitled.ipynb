```python
import mysql.connector
```

```python
def get_connection():
    return mysql.connector.connect(
        host='localhost',
        user='root',
        password='123456',
        database='dbms'
    )
```

```python
def insert_students(id,name,age):
    db=get_connection();
    cursor=db.cursor();
    query='INSERT INTO students VALUES(%s,%s,%s)';
    cursor.execute(query,(id,name,age));
    print(f"INSERTED id:{id},name:{name},age:{age}");
    db.commit();
    cursor.close();
    db.close();
```

```python
def update_students(id,new_name,new_age):
    db=get_connection();
    cursor=db.cursor();
    query='UPDATE students SET name=%s,age=%s WHERE id=%s';
    cursor.execute(query,(new_name,new_age,id));
    print(f"UPDATED id:{id},name:{new_name},age:{new_age}");
    db.commit();
    cursor.close();
    db.close();
```

```python
def delete_students(id):
    db=get_connection();
    cursor=db.cursor();
    query='DELETE  from students WHERE id=%s';
    cursor.execute(query,(id,));
    print(f"DELETE id:{id}");
    db.commit();
    cursor.close();
    db.close();
```

```python
while(True):
    ch=int(input("Enter choice (1->insert \n2->Update\n3->delete\n4->END\n)"));
    if(ch==1):
        id=int(input("Enter Id"));
        name=input("Name:");
        age=int(input("Age:"));
        insert_students(id,name,age);
    elif(ch==2):
        id=int(input("Enter Id"));
        name=input("Name:");
        age=int(input("Age:"));
        update_students(id,name,age);
    elif(ch==3):
       id=int(input("Enter Id"));
       delete_students(id);
    elif(ch==4):
        print("Exitinggg.....");
        break;
    else:
        print("Invalid Choice");
        
```
